# EDA on multivariate data

### 1. Histplot

```python
# sns.histplot(data=titanic, x='age', hue='sex')
plt.figure()
sns.kdeplot(titanic['age'])
plt.show()
# this plots the PDF (probability density function wrt age), so it means, if age is 20, density is 0.015,
# and hence the probability of a person being alive (target) of age 20 would be 0.15
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', style='sex', size='fare' )
```

```python
sns.barplot(data=titanic, x='pclass', y='age', hue='sex')
```
